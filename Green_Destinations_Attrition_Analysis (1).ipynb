# Project Brief: Green Destinations

### Objective

The HR Director wants to analyze employee attrition trends at Green Destinations.

### Step 1: Load Data and Explore

```python
import pandas as pd

data = pd.read_csv('greendestination.csv')
```

### Step 2: Calculate Attrition Rate

```python
# Calculate attrition rate
attrition_rate = data['Attrition'].value_counts(normalize=True)['Yes'] * 100
attrition_rate
```

### Step 3: Analyze Factors Influencing Attrition

```python
# Convert 'Attrition' to a binary format for analysis (1 for 'Yes', 0 for 'No')
data['AttritionBinary'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Analyze Age, YearsAtCompany, and DailyRate by Attrition
attrition_analysis = data.groupby('AttritionBinary')[['Age', 'YearsAtCompany', 'DailyRate']].mean()
attrition_analysis
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Create a figure for the plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Distribution of Attrition
sns.countplot(x='Attrition', data=data, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Attrition')

# Plot 2: Boxplot for Age by Attrition
sns.boxplot(x='Attrition', y='Age', data=data, ax=axes[0, 1])
axes[0, 1].set_title('Age Distribution by Attrition')

# Plot 3: Boxplot for YearsAtCompany by Attrition
sns.boxplot(x='Attrition', y='YearsAtCompany', data=data, ax=axes[1, 0])
axes[1, 0].set_title('Years at Company by Attrition')

# Plot 4: Boxplot for DailyRate by Attrition
sns.boxplot(x='Attrition', y='DailyRate', data=data, ax=axes[1, 1])
axes[1, 1].set_title('Income (DailyRate) by Attrition')

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

### Conclusion
- The attrition rate is **16.12%**.
- Employees who left are slightly younger and have fewer years at the company.
- There is a small difference in income between employees who stayed and those who left.
